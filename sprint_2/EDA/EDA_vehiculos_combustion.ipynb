```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('vehiculos_combustion_CO2_2023.csv')
```

```python
print(data.head())
```

```python
##Calculo las descriptivas
print(data.describe())
```

```python
print(data.dtypes)
```

```python
#Verifico los valores nulos
print(data.isnull().sum())
```

```python
# Histograma del tamao del motor
#Con este histograma puedo observar como se distribuye el tamao del motor teniendo en cuenta su frecuencia
plt.figure(figsize=(10, 6))
sns.histplot(data['Engine Size(L)'], bins=20, kde=True)
plt.title('Distribucin del Tamao del Motor')
plt.xlabel('Tamao del Motor (L)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Grfico de dispersin entre tamao del motor y emisiones de CO2
#En este grfico puedo observa la relacin marcada que existe entre el tamao del motor y las emisiones de CO2.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size(L)', y='CO2 Emissions(g/km)', data=data)
plt.title('Relacin entre Tamao del Motor y Emisiones de CO2')
plt.xlabel('Tamao del Motor (L)')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.show()
```

```python
# Boxplot por Clase de Vehculo
#En este grafico puedo observar la distribucin de las emisiones de CO2 por clase de vehculo. Tambien se pueden observar los outliers.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Vehicle Class', y='CO2 Emissions(g/km)', data=data)
plt.title('Distribucin de Emisiones de CO2 por Clase de Vehculo')
plt.xlabel('Clase de Vehculo')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Matriz de Correlacin
#En esta matriz de correlacin puedo observar la relacin entre las distintas variables numricas. Se pueden observar varias correlaciones bien marcadas.
correlation_matrix = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlacin')
plt.show()
```

```python
# Distribucin de emisiones de CO2 por tipo de combustible
#En este grfico se puede observar la distribucin de las emisiones de CO2 por tipo de combustible. Adems se pueden visualizar los outliers.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel(Type)', y='CO2 Emissions(g/km)', data=data)
plt.title('Distribucin de Emisiones de CO2 por Tipo de Combustible')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.xticks(rotation=45)
plt.show()
```
