```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('vehiculos_combustion_CO2_2023.csv')
```

```python
print(data.head())
```

```python
##Calculo las descriptivas
print(data.describe())
```

```python
print(data.dtypes)
```

```python
#Verifico los valores nulos
print(data.isnull().sum())
```

```python
# Histograma del tamaño del motor
#Con este histograma puedo observar como se distribuye el tamaño del motor teniendo en cuenta su frecuencia
plt.figure(figsize=(10, 6))
sns.histplot(data['Engine Size(L)'], bins=20, kde=True)
plt.title('Distribución del Tamaño del Motor')
plt.xlabel('Tamaño del Motor (L)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Gráfico de dispersión entre tamaño del motor y emisiones de CO2
#En este gráfico puedo observa la relación marcada que existe entre el tamaño del motor y las emisiones de CO2.
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Engine Size(L)', y='CO2 Emissions(g/km)', data=data)
plt.title('Relación entre Tamaño del Motor y Emisiones de CO2')
plt.xlabel('Tamaño del Motor (L)')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.show()
```

```python
# Boxplot por Clase de Vehículo
#En este grafico puedo observar la distribución de las emisiones de CO2 por clase de vehículo. Tambien se pueden observar los outliers.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Vehicle Class', y='CO2 Emissions(g/km)', data=data)
plt.title('Distribución de Emisiones de CO2 por Clase de Vehículo')
plt.xlabel('Clase de Vehículo')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Matriz de Correlación
#En esta matriz de correlación puedo observar la relación entre las distintas variables numéricas. Se pueden observar varias correlaciones bien marcadas.
correlation_matrix = data.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')
plt.show()
```

```python
# Distribución de emisiones de CO2 por tipo de combustible
#En este gráfico se puede observar la distribución de las emisiones de CO2 por tipo de combustible. Además se pueden visualizar los outliers.
plt.figure(figsize=(10, 6))
sns.boxplot(x='Fuel(Type)', y='CO2 Emissions(g/km)', data=data)
plt.title('Distribución de Emisiones de CO2 por Tipo de Combustible')
plt.xlabel('Tipo de Combustible')
plt.ylabel('Emisiones de CO2 (g/km)')
plt.xticks(rotation=45)
plt.show()
```
