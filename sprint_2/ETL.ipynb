#### estacion de carga electrica y de combustion

```python
#cargamos el dataset estacion de carga electrica y de combustion
est_elec_comb = pd.read_csv('../Datasets/Electric and Alternative Fuel Charging Stations.csv')
```

```python
#observamos numero de filas, columnas y las primeras 5 filas del dataset
print(est_elec_comb.shape)
est_elec_comb.head(5)
```

```python
# observamos nombre de columnas, cantidad de valores no nulos y tipo de datos
est_elec_comb.info()
```

se escogieron 17 columnas del dataset sobre las estaciones electricas y de combustion las cuales por calidad de los datos.

**Diccionario de datos** de las columnas seleccionadas

- 'ID': Identificación - Un identificador único para la estación.
- 'Fuel Type Code': Código de Tipo de Combustible
- 'Station Name': Nombre de la Estación
- 'Street Address': Dirección de Calle 
- 'City': Ciudad - La ciudad donde se encuentra la estación.
- 'State': Estado - El estado donde se encuentra la estación.
- 'Country': País - El país donde se encuentra la estación.
- 'Latitude': Latitud - La latitud geográfica de la ubicación de la estación.
- 'Longitude': Longitud - La longitud geográfica de la ubicación de la estación.
- ZIP': Código ZIP - El código postal de la ubicación de la estación.
- 'Status Code': Código de Estado - El estado de funcionamiento de la estación
- 'Groups With Access Code': Grupos con Código de Acceso - Los grupos que tienen acceso a la estación.
- 'Access Days Time': Horario de Días de Acceso
- EV Network': Red de Carga EV - La red de carga de vehículos eléctricos asociada a la estación.
- 'Geocode Status': Estado de Geocodificación - El estado del proceso de geocodificación de la ubicación.
- 'Date Last Confirmed': Fecha de Última Confirmación - La fecha en que la información fue confirmada por última vez.
- 'Updated At': Actualizado en - La fecha de la última actualización de la información

```python
## Seleccionar las columnas deseadas para el nuevo DataFrame

columnas = [
    'ID','Fuel Type Code', 'Station Name','Street Address', 'City', 'State','Country', 'Latitude', 'Longitude','ZIP',
    'Status Code', 'Groups With Access Code', 'Access Days Time', 'EV Network','Geocode Status','Updated At'
]

estacion = est_elec_comb[columnas]

# Imprimir el nuevo DataFrame
estacion.head(2)
```

```python
# cantidad de filas y columnas
estacion.shape
```

```python
# observamos nombre de columnas, cantidad de valores no nulos y tipo de datos
estacion.info()
```

```python
# reemplazamos los valores nulos como no especificado
estacion.fillna('no especificado', inplace = True)
```

```python
estacion.info()
```

```python
# exportamos
#exportamos el archivo con la primera transformacion
estacion.to_csv('Datasets_limpios/estacion.csv',index=False)
```

#### vehiculos de combustion

```python
#cargamos el dataset vehiculos de combustion
veh_com = pd.read_csv('../Datasets/Vehicle Fuel Economy Data.csv')
```

```python
#observamos numero de filas, columnas y las primeras 5 filas del dataset
print(veh_com.shape)
veh_com.head(3)
```

```python
# observamos nombre de columnas, cantidad de valores no nulos y tipo de datos
veh_com.info()
```

```python
# Mover la columna 'ColumnaC' a la primera posición
columna_movida = veh_com.pop('id')  # Extraer la columna
veh_com.insert(0, 'id', columna_movida)  # Insertarla en la primera pid
```

```python
#cantidad de valores duplicados
veh_com['id'].duplicated().sum()
```

```python
veh_com.drop_duplicates(subset=['id'], keep='first', inplace=True)
```

```python
# Seleccionar las columnas con un total de registros mayor a 44169
veh_com = veh_com.loc[:, veh_com.count() > 44169]
```

```python
#borra la fila con valores faltantes
veh_com.dropna(inplace = True)
```

```python
veh_com['id'] = veh_com['id'].astype(int)
```

```python
veh_com.info()
```

```python
veh_com.head()
```
